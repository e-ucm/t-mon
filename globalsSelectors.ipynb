**Style** widgets to select the style to display all vis with this style

```python
global style
stylelist=['default', 'classic', 'dark_background', 'grayscale', 'seaborn', 'ggplot', 'fivethirtyeight', 'Solarize_Light2']
style=widgets.Dropdown(options=stylelist,
                       description='Plot style',
                       disabled=False)

def change_style_use(change):
    plt.style.use(change['new'])
    on_selected_tab_index_change({'old': tabs.selected_index, 'new': tabs.selected_index})
style.observe(change_style_use, 'value')
style.value='default'
```

***on_selected_tab_index_change* function** display the visualisation in the tab selected and clear old vis display in the previous tab

Input : change : dictionary {'old' : previous_index, 'new' : index selected}

```python
def on_selected_tab_index_change(change):
        # Enabling the `widget` backend.
        # This requires jupyter-matplotlib a.k.a. ipympl.
        # ipympl can be install via pip or conda.
        %matplotlib widget

        old=change['old']
        if old==0:
            closeFigGameSelected()
        elif old==1:
            closeFigVideos()
        elif old==2:
            closeFigProgress()
        elif old==3:
            closeFigCompletable()
        elif old==4:
            closeFigAlternative()
        elif old==5:
            closeFigInteractions()
        elif old==6:
            closeFigAccessibles()
        else : #old==7
            closeFigMenus()
        new=change['new']
        if new==0:
            #Pie chart with number of games started and completed
            %run vis/xAPISG-GamesStartedCompleted.ipynb
            tabs.children[0].children[1].children[1].children=vis_games_started_completed()
        elif new==1:
            #Bar chart with number of times each video was seen and skipped
            %run vis/xAPISG-VideosSeenSkipped.ipynb
            tabs.children[1].children[1].children[1].children=vis_videos_seen_skipped()
        elif new==2:
            #Line chart with progress of players per time
            %run vis/xAPISG-PlayersProgress.ipynb
            tabs.children[2].children[1].children[1].children=vis_players_progress()
        elif new==3:
            #Line chart with progress of players in each completable
            %run vis/xAPISG-CompletableProgressIncreaseDecrease.ipynb
            tabs.children[3].children[1].children[1].children=vis_completables_progress_increase_decrease()
            #Bar chart with progress of players in each completable
            %run vis/xAPISG-CompletablesProgress.ipynb
            tabs.children[3].children[2].children[1].children=vis_completables_progress()
            #Bar chart with scores of players in each completable
            %run vis/xAPISG-CompletablesScores.ipynb
            tabs.children[3].children[3].children[1].children=vis_completables_scores()
            #Bar chart with max and min completion time in each completable
            %run vis/xAPISG-CompletablesTimes.ipynb
            tabs.children[3].children[4].children[1].children=vis_completables_times()
        elif new==4:
            #Bar chart with number of correct and incorrect answers per player
            %run vis/xAPISG-CorrectIncorrectPerPlayer.ipynb
            tabs.children[4].children[1].children[1].children=vis_correct_incorrect_player()
            #Bar chart with number of correct and incorrect answers per question (alternative)
            %run vis/xAPISG-CorrectIncorrectPerQuestion.ipynb
            tabs.children[4].children[2].children[1].children=vis_correct_incorrect_question()
            #Bar chart per question, with number of times each alternative has been selected
            %run vis/xAPISG-AlternativesSelectedQuestion.ipynb
            tabs.children[4].children[3].children[1].children=vis_alternatives_selected_question()
        elif new==5:
            #item interacted
            %run vis/xAPISG-ItemsInteracted.ipynb
            tabs.children[5].children[1].children[1].children=vis_items_interacted_function_time_interacted()
            tabs.children[5].children[2].children[1].children=vis_items_interacted_times_heatmap_interacted()
            tabs.children[5].children[3].children[1].children=vis_items_interacted_times_interact()
            %run vis/xAPISG-ItemsActionTypeInteracted.ipynb
            tabs.children[5].children[4].children[1].children=vis_items_action_type_interacted_times_interact()
        elif new==6:
            #Accessibles
            %run vis/xAPISG-AccessedAccessible.ipynb
            tabs.children[6].children[1].children[1].children=vis_accessibles_function_time_interacted()
            tabs.children[6].children[2].children[1].children=vis_accessibles_times_heatMap_interacted()
            tabs.children[6].children[3].children[1].children=vis_accessibles_times_interacted()
        else : #new==7
            #menus selected
            %run vis/xAPISG-MenusSelected.ipynb
            tabs.children[7].children[1].children[1].children=vis_menu_selected_interactive()
```

***display_checkboxes* function** is constructing the multicheckboxes widgets for all multis selections

```python
%run vis/helpersFunctions/getterObjectsList.ipynb
def display_checkboxes():
    #Globals selectors
    %run widgets/multipleSelectorCheckboxesWidget.ipynb
    class MultiCheckboxWidgetUpdateAllVis(MultiCheckboxWidget):
        def update(self):
            on_selected_tab_index_change({'old': tabs.selected_index, 'new': tabs.selected_index})
    global checkboxesPlayersSelected
    players_list=sorted(list(players_info.keys()))
    checkboxesPlayersSelected=MultiCheckboxWidgetUpdateAllVis("Players", players_list)
    global checkboxesVideoSelected
    videos_list=get_full_list_of_objects("videos_seen",action2="videos_seen", dict=False)
    checkboxesVideoSelected=MultiCheckboxWidgetUpdateAllVis("Videos", videos_list)
    global checkboxesAlternativeSelected
    alternatives_list= get_full_list_of_objects("alternatives")
    checkboxesAlternativeSelected=MultiCheckboxWidgetUpdateAllVis("Alternative", alternatives_list)
    global checkboxesCompletableSelected,completable_score_list, completable_progress_list,\
    completables_times_list, completables_increase_decrease_list
    completable_score_list=get_full_list_of_objects("completables_scores")
    completable_progress_list, completables_increase_decrease_list=getlist_completable_progress()
    completables_times_list=get_full_list_of_objects("completables_times")
    completables_list=set()
    completables_list=completables_list.union(completable_score_list)
    completables_list=completables_list.union(completable_progress_list)
    completables_list=completables_list.union(completables_increase_decrease_list)
    completables_list=completables_list.union(completables_times_list)
    completables_list=list(completables_list)
    checkboxesCompletableSelected=MultiCheckboxWidgetUpdateAllVis("Completables",completables_list)
    global checkboxesItemInteractedSelected, item_interaction_action_type_list, item_interaction_list
    items=set()
    item_interaction_action_type_list=get_full_list_of_objects("action_type_interaction", True)
    items=items.union(item_interaction_action_type_list)
    item_interaction_list=get_full_list_of_objects("interactions", True)
    items=items.union(item_interaction_list)
    items=list(items)
    checkboxesItemInteractedSelected=MultiCheckboxWidgetUpdateAllVis("Items interactions",items)
    global checkboxesAccessibleSelected
    accessibles=get_full_list_of_objects("accessible")
    checkboxesAccessibleSelected=MultiCheckboxWidgetUpdateAllVis("Accessibles",accessibles)
    global checkboxesSelectedmenus
    menus=get_full_list_of_objects("selected_menus")
    checkboxesSelectedmenus=MultiCheckboxWidgetUpdateAllVis("Menus selected",menus)
```
